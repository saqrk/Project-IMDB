Q3. Find out the unique genres of movies in this dataset.

```python
df.info()
```

```python
gen_df = df[~df['genres'].isna()]

gen_lst = []

for index, row in gen_df.iterrows():
    genre = gen_df.loc[index, 'genres']
    for k in genre:
        gen_lst.append(k['name'])
        
pd.DataFrame(set(gen_lst), columns=["Unique Genres"])
```

```python
gen_df = pd.DataFrame(df.loc[:,'genres'])

gen_lst = []

for index, row in gen_df.iterrows():
    genre = gen_df.loc[index, 'genres']
    if isinstance(genre, list):
        for k in genre:
            gen_lst.append(k['name'])
        
pd.DataFrame(set(gen_lst), columns=["Unique Genres"])
```

Q4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
crew_df = df[~df['crew'].isna()]

def prod_lst(index):
    movie = crew_df.iloc[index]
    crew_lst = movie.loc['crew']
    producer_lst = []
    for elem in crew_lst:
        if elem['job'] == 'Producer':
            producer_lst.append(elem['name'])
    return producer_lst
```

```python
prod_lst(20)
```

```python
def direct_lst(index):
    movie = crew_df.iloc[index]
    crew_lst= movie.loc['crew']
    director_lst = []
    for elem in crew_lst :
        if elem['job'] == 'Producer':
            director_lst.append(elem['name'])
    return director_lst
```

```python
direct_lst(20)
```

```python
table = pd.DataFrame(columns = ['Movie Title', 'Producers','Director','ROI'])
```

```python
for index, row in crew_df.iterrows():
    try:
        table = table.append({'Movie Title':crew_df.loc[index, 'original_title'],
                            'Producers':prod_lst(index),
                             'Director':direct_lst(index),
                             'ROI':crew_df.loc[index,'roi']},
                            ignore_index = True)
        
    except:
        continue
```
